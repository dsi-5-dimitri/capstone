```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np
import os
import re
import sys
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sub = ['bitcoincash', 'bitcoin', 'ethereum', 'vechain', 'eos', 'ripple'] 
csv1, csv2, csv3, csv4 = ([] for i in range(4))
for i in sub:
    csv1.append(i + '_controversial.csv')
    csv2.append(i + '_hot.csv')
    csv3.append(i + '_new.csv')
    csv4.append(i + '_top.csv')
```

```python
# for i in sub[-2:-3]:
#     os.mkdir('./' + i.lower())
```

```python
csvx = csv1 + csv2 + csv3 + csv4
loc, got, com, dir1 = ([] for i in range(4))

for h in range(5):
    for i in sub:
        dir1.append('./' + i.lower())  
for (i, j) in zip(csvx, dir1):
    com.append(j + '/' + i[:-4].lower() + '_sentiment.csv')
    loc.append('./s1/' + i[:-4] + '_comments.csv')
    got.append('./data/' + i)
```

```python
ix1, neg, neu, pos, comp, drop = ([] for i in range(6))
ix2, ix3, ix4 = (0 for i in range(3))
analyzer = SentimentIntensityAnalyzer()
pattern = re.compile(r'^[a-zA-Z]+_[a-zA-Z]+')

ix2 = 0
for i in loc:
    df = pd.read_csv(i)
    print(df.shape)
#     dfx = pd.read_csv(j)
#     print(dfx.shape)
#     df = pd.merge(df, dfx, left_index = True, right_index = True)
    x = pattern.findall(df.columns.values[1])[0].lower()
    df.columns.values[1] = x
    y = df[df[x].isnull()].index
    for i in y.tolist():
        ix4 += 1
        drop.append(i)
        df.drop(i, inplace = True)
    z = list(df[x].values)
    for i in z:
        vs = analyzer.polarity_scores(i)
        ix1.append(z.index(i))    
        neg.append(vs['neg'])
        neu.append(vs['neu'])
        pos.append(vs['pos'])
        comp.append(vs['compound'])
        ix3 += 1
        sys.stdout.write(f'\rix3: {ix3} dropped: {ix4}, {drop}')
    sent = pd.DataFrame({'ix1' : ix1, 'negative' : neg, 'pos' : pos, 'neu' : neu, 'comp' : comp})
    df = pd.merge(df, sent, left_index = True, right_index = True)
    df.to_csv(com[ix2], columns = df, index = None)
    print(f'\nsaving file as: {com[ix2]}\n')
    ix2 += 1
    drop = []
    ix3, ix4 = (0 for i in range(2))
```

```python
ix = 0
for i, j in zip(got, com):
    dfx1 = pd.read_csv(i)
    dfx2 = pd.read_csv(j)
    print(f'dropped: {dfx2.ids.isnull().sum()}')
    dfx2.dropna(inplace = True)
    ab = dfx1.set_index('ids')['num_comments'].to_dict()
    place = []
    for k in dfx2.ids:
        place.append(ab[k])
    dfx3 = pd.DataFrame({'num' : place})
    dfx2['num'] = dfx3
    print(f'saved as: {com[ix][:-4]}1')
    sav = com[ix][:-4] + '1.csv'
    dfx2.to_csv(sav, columns = dfx2, index = None)
    ix += 1
```
